```python
gold_df = pd.read_parquet('gold//tariff_gold.parquet')
engine = get_engine()

def test_connection(engine):
    try: 
        with engine.connect() as connection: 
            print("Connection successful")
    except Exception as e:
        print(f"Connection failed: {e}")

def load_data_to_sql(df, table_name, engine):
    try:
        df.to_sql(table_name, con=engine, if_exists='replace', index=False)
        print(f"Data loaded successfully into {table_name}")
    except Exception as e:
        print(f"Failed to load data: {e}")

```

```python
test_connection(engine)
load_data_to_sql(gold_df, 'yearly_tariffs', engine)
verify_df = pd.read_sql("SELECT TOP 5 * FROM yearly_tariffs", con=engine)
print(verify_df)
```
